### Descripcion de variables:

##### track_id: 
El identificador de Spotify para la canción.

##### artists: 
Los nombres de los artistas que interpretaron la canción. Si hay más de un artista, se separan por punto y coma (;).

##### album_name: 
El nombre del álbum en el que aparece la canción.

##### track_name: 
nombre de la canción.

##### popularity: 
La popularidad de una canción, medida en una escala de 0 a 100, donde 100 representa la máxima popularidad. Se calcula mediante un algoritmo basado en el número total de reproducciones de la canción y cuán recientes son esas reproducciones. Las canciones populares actualmente tendrán una mayor puntuación. Variable a predecir

##### duration_ms:
La duración de la canción en milisegundos.

##### explicit: 
Indica si la canción contiene letras explícitas (true = sí lo tiene; false = no lo tiene O desconocido).

##### danceability: 
La capacidad de la canción para ser adecuada para el baile, basada en elementos musicales como el tempo, la estabilidad del ritmo, la fuerza del ritmo y la regularidad general. Un valor de 0.0 es la menos adecuada para bailar, y 1.0 es la más adecuada.

##### energy: 
La energía de la canción, medida en una escala de 0.0 a 1.0, que representa la percepción de intensidad y actividad. Las canciones enérgicas se sienten generalmente rápidas, ruidosas y con mucha actividad.

##### key:
La tonalidad de la canción, mapeada a notas mediante la notación estándar de Clase de Tonalidad. Por ejemplo, 0 = C, 1 = C♯/D♭, 2 = D, y así sucesivamente. Si no se detecta ninguna tonalidad, el valor es -1.

##### loudness: 
El volumen general de la canción en decibelios (dB).

##### mode: 
El modo, que indica la modalidad (mayor o menor) de la canción. Mayor se representa como 1 y menor como 0.

##### speechiness: 
probabilidad de que la canción contenga palabras habladas. Valores cercanos a 1.0 indican una alta probabilidad de discurso, mientras que valores entre 0.33 y 0.66 pueden representar música y discurso combinados.

##### acousticness: 
Una medida de confianza de si la canción es acústica, con 1.0 indicando alta confianza en que es acústica.

##### instrumentalness:
Predice si la canción no contiene vocales. Valores cercanos a 1.0 indican alta probabilidad de contenido instrumental.

##### liveness:
Detecta la presencia de una audiencia en la grabación. Valores altos indican una alta probabilidad de que la canción se interpretara en vivo.

##### valence: 
Una medida de la positividad musical transmitida por la canción. Valores altos indican positividad (p. ej., felicidad), mientras que valores bajos indican negatividad (p. ej., tristeza).

##### tempo:
El tempo estimado de la canción en pulsos por minuto (BPM).

##### time_signature: 
Una estimación de la firma de tiempo, que indica cuántos pulsos hay en cada compás musical (o medida). Los valores varían de 3 a 7.

##### track_genre:
El género al que pertenece la canción

#### Top 10 artistas mas populares:
Analizamos el DataFrame  para identificar a los artistas más populares. Considerando múltiples artistas por canción, calculamos el promedio de popularidad para cada uno. La tabla resultante presenta los 10 artistas con los promedios más altos, brindando una visión concisa de los artistas más destacados en términos de popularidad en nuestro conjunto de datos musical

```python
# Filtrar canciones con popularidad mayor a 70 en df3
df3_populares = df3[df3['Popularidad'] >= 70]

# Dividir los artistas separados por ;
df3_populares['Artistas'] = df3_populares['Artistas'].str.split(';')

# Expandir las listas de artistas en nuevas filas
df3_populares_expandido = df3_populares.explode('Artistas')

# Contar la cantidad de canciones por artista
canciones_por_artista = df3_populares_expandido['Artistas'].value_counts()

# Filtrar artistas con más de 3 canciones con popularidad mayor a 70
artistas_calificables = canciones_por_artista[canciones_por_artista > 3].index

# Filtrar el DataFrame original por los artistas calificables en df3
df3_calificable = df3_populares_expandido[df3_populares_expandido['Artistas'].isin(artistas_calificables)]

# Calcular el promedio de la popularidad por artista en df3
artistas_promedio_popularidad_df3 = df3_calificable.groupby('Artistas')['Popularidad'].mean().round(1)

# Ordenar los artistas por su promedio de popularidad en df3 en orden descendente
artistas_promedio_popularidad_df3 = artistas_promedio_popularidad_df3.sort_values(ascending=False)

# Seleccionar los 10 artistas con el promedio de popularidad más alto en df3
top_10_artistas_promedio_df3 = artistas_promedio_popularidad_df3.head(10)

# Crear un nuevo DataFrame con los resultados en df3
tabla_top_artistas_promedio_df3 = pd.DataFrame({
    "Artista": top_10_artistas_promedio_df3.index,
    "Promedio de Popularidad": top_10_artistas_promedio_df3.values
})

# Mostrar la tabla de los 10 artistas con el promedio de popularidad más alto en df3
print(tabla_top_artistas_promedio_df3)
```

### Pregunta de investigacion 1:¿Existe una correlación entre la popularidad de una canción y su género musical?

#### Hipótesis nula (H0):
No hay diferencia significativa en la popularidad promedio entre los géneros musicales.
#### Hipótesis alternativa (H1):
Existe una diferencia significativa en la popularidad promedio entre al menos dos géneros musicales.

```python
# Contar la cantidad de canciones por género y seleccionar los 20 más comunes
top_20_generos = df3['Género_pista'].value_counts().nlargest(20)

# Crear el gráfico de barras
plt.figure(figsize=(12, 8))
top_20_generos.plot(kind='bar', color='skyblue')
plt.title('Top 20 Géneros Musicales con Más Canciones')
plt.xlabel('Género Musical')
plt.ylabel('Cantidad de Canciones')
plt.xticks(rotation=45, ha='right')
plt.show()

# Generar una tabla que muestra la cantidad de canciones con popularidad mayor o igual a 70 por género
tabla_popularidad = df3[df3['Popularidad'] >= 70].groupby('Género_pista').size().reindex(top_20_generos.index)
print(tabla_popularidad)
```

El análisis de los resultados revela que, entre los 20 géneros musicales más frecuentes en el conjunto de datos, algunos de ellos no tienen canciones con una popularidad igual o superior a 70. Los géneros presentan un valor NaN en la tabla de conteo de canciones con alta popularidad. Esto sugiere que estas categorías podrían no ser representativas en términos de popularidad elevada según el conjunto de datos proporcionado.

A continuación, presentamos un resumen estadístico de géneros musicales, excluyendo aquellos que no cuentan con canciones de alta popularidad (clasificadas con un valor de 85 o superior). Este análisis busca destacar las características de la popularidad en los géneros seleccionados, proporcionando una visión detallada de su distribución y variabilidad.

```python
# Filtrar géneros cuyo valor máximo de popularidad sea mayor o igual a 85
generos_con_popularidad_85 = df3.groupby('Género_pista')['Popularidad'].max()
generos_filtrados = generos_con_popularidad_85[generos_con_popularidad_85 >= 85].index
df_filtrado = df3[df3['Género_pista'].isin(generos_filtrados)]

# Calcular estadísticas descriptivas por género
estadisticas_por_genero = df_filtrado.groupby('Género_pista')['Popularidad'].describe()

# Agregar la mediana y el rango intercuartílico (IQR)
estadisticas_por_genero['Mediana'] = df_filtrado.groupby('Género_pista')['Popularidad'].median()
estadisticas_por_genero['IQR'] = estadisticas_por_genero['75%'] - estadisticas_por_genero['25%']

# Ordenar por el tercer cuartil (Q3) de mayor a menor
estadisticas_ordenadas = estadisticas_por_genero.sort_values(by='75%', ascending=False)

# Formatear los números a un decimal
estadisticas_ordenadas = estadisticas_ordenadas.round(1)

# Configuración de estilo de seaborn (opcional)
sns.set(style="whitegrid")

# Crear el boxplot ordenado por Q3 de mayor a menor
plt.figure(figsize=(12, 8))
ax = sns.boxplot(x='Género_pista', y='Popularidad', data=df_filtrado, order=estadisticas_ordenadas.index, palette="Set3")

# Rotar las etiquetas del eje x para mejorar la legibilidad
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Agregar título y etiquetas
plt.title('Boxplot de Popularidad por Género de Pista (Ordenado por Q3)')
plt.xlabel('Género de Pista')
plt.ylabel('Popularidad')

# Mostrar el boxplot
plt.legend()  # Agregar leyenda para la línea horizontal
plt.tight_layout()  # Ajustar el diseño para mejorar la visualización
plt.show()

# Imprimir la tabla de estadísticas ordenada por Q3
print(estadisticas_ordenadas[['count', 'mean', 'std', 'min', '25%', 'Mediana', '75%', 'max', 'IQR']])
```

El análisis de los géneros musicales revela una amplia variabilidad en la popularidad de las canciones. Se observa que algunos géneros, como "dance", "rock" y "alternative", exhiben una dispersión significativa, indicada por el alto valor del rango intercuartílico (IQR). Por otro lado, géneros como "reggae", "reggaeton" y "latino" presentan una alta variabilidad y una mediana más baja.

El siguiente gráfico de torta ilustra los géneros predominantes de las canciones más populares en Spotify. Intenta mostrar cuales generos poseen mayor canciones cuya popularidad es alta.

```python
# Filtrar las canciones con popularidad mayor a 70
canciones_populares = df3[df3['Popularidad'] >= 70]

# Cuenta cuántas canciones hay en cada género
conteo_por_genero = canciones_populares['Género_pista'].value_counts()

# Calcular porcentaje de popularidad en cada categoría
porcentaje_por_categoria = (conteo_por_genero / conteo_por_genero.sum()) * 100

# Filtrar las primeras diez categorías
top_10_categorias = porcentaje_por_categoria.head(10)

# Crear un gráfico de torta
fig, ax = plt.subplots(figsize=(12, 12))

# Dibujar un gráfico de torta
wedges, texts, autotexts = ax.pie(top_10_categorias, autopct='%1.1f%%', startangle=140)

# Agregar línea entre los géneros
for w in wedges:
    w.set_linewidth(2)
    w.set_edgecolor('white')

# Configuración de etiquetas
labels = top_10_categorias.index
ax.legend(wedges, labels, title="Géneros", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=12)

# Título y mostrar el gráfico
plt.title('Canciones Populares (Popularidad >= 70) por Género')
plt.axis('equal')

plt.show()
```

Este género presenta una clara predominancia de canciones con niveles de popularidad más altos en comparación con otros géneros. #### Análisis Estadístico:

Se realizó un análisis de regresión lineal ordinaria (OLS) para predecir la popularidad de las canciones basándose en varias características. Aquí se presentan los resultados clave:

R-cuadrado: Indica la proporción de variabilidad en la variable dependiente (popularidad) que es explicada por el modelo. En este caso, el R-cuadrado es 0.060, lo que significa que aproximadamente el 6% de la variabilidad en la popularidad se explica por las variables independientes en el modelo.

F-estadística: Evalúa si al menos una de las variables independientes tiene un efecto significativo en la variable dependiente. La F-estadística es 627.4 con un valor p de 0.00, lo que sugiere que al menos una variable independiente tiene un efecto significativo en la popularidad.

Coeficientes: Los coeficientes asociados con cada variable independiente indican la magnitud y dirección de su efecto en la variable dependiente (popularidad).

Errores estándar: Indican la precisión de las estimaciones de los coeficientes. Un error estándar más pequeño sugiere una estimación más precisa.

Intervalos de confianza: Proporcionan un rango dentro del cual se espera que esté el verdadero valor del coeficiente con un determinado nivel de confianza.

#### Resumen:

El modelo en su conjunto tiene un bajo poder explicativo (R-cuadrado de 0.060), lo que indica que la variabilidad en la popularidad no se explica bien con las variables incluidas en el modelo.

Las variables significativas incluyen bailabilidad, energía, volumen, hablado, acusticidad, instrumentalidad, valencia, tempo y firma de tiempo.

Las variables de tonalidad no son estadísticamente significativas, ya que sus valores p (p>|t|) son mayores que 0.05.

La falta de significancia estadística para la tonalidad sugiere que, según este modelo, la tonalidad específica de una canción no tiene un efecto significativo en su popularidad.

### Pregunta de investigacion 5:¿Hay una relación entre la duración de una canción y su popularidad?

#### Hipótesis nula (H0):
La duración de una canción no tiene influencia en su popularidad.
#### Hipótesis alternativa (H1): 
La duración de una canción tiene influencia en su popularidad.

```python
# Crear un gráfico de dispersión con regresión lineal
plt.figure(figsize=(10, 6))
sns.regplot(data=df3, x='Duración_s', y='Popularidad', scatter_kws={'alpha': 0.3}, line_kws={'color': 'orange'})

# Etiquetas de ejes y título
plt.xlabel('Duración de la Canción (segundos)')
plt.ylabel('Popularidad')
plt.title('Relación entre Duración de la Canción y Popularidad')

# Ajustar el modelo de regresión lineal
X = df3['Duración_s']
X = sm.add_constant(X)  # Agregar una constante para el término independiente
y = df3['Popularidad']
modelo = sm.OLS(y, X).fit()

# Calcular el R cuadrado
r_cuadrado = modelo.rsquared
print(f"R cuadrado: {r_cuadrado:.3f}")

# Mostrar el gráfico
plt.show()
```
